## Before we finish - the most important thing
#### 'End of session' routine. Good practice - run this set of commands regularly for back up and version control.
* Open terminal
* `git status` - to see commit status (check uncommitted changes to your project)
* Clear all cell output (there is a corresponding button above)
* Update .gitignore file (more on it later...)
* `git add --all`
* `git commit -m "commit message"`
* `git push`
* Refresh your GitHub repository page to check that all changes are committed and pushed. Well done - you successfully committed your code to GitHub.
