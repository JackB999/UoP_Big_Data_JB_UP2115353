### Exercise 1

```python
# 1) Ask user to enter a number
n = int(input("Enter a number: "))

# 2) Check if the number is prime
if n <= 1:
    print("This is not a prime number")
else:
    is_prime = True

    for i in range(2, n):
        if n % i == 0:
            is_prime = False

    # 3) Print the result
    if is_prime:
        print("This is a prime number")
    else:
        print("This is not a prime number")
```

### Exercise 2

```python
# Ask user how many Fibonacci numbers to display
n = int(input("How many Fibonacci numbers? "))

a = 0
b = 1

# Print Fibonacci sequence
for i in range(n):
    print(a, end=' ')
    next_num = a + b
    a = b
    b = next_num
```

### Exercise 3

```python
# Print the pattern (decreasing triangle)
n = 5

for i in range(n, 0, -1):
    for j in range(i):
        print('*', end=' ')
    print()
```
