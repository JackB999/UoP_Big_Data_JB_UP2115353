Code from lecture 1.2 - 2.1

```python
# 1) Ask user to enter a number
n = int(input("Enter a number: "))

# 2) Check if the number is prime
if n <= 1:
    print("This is not a prime number")
else:
    is_prime = True

    for i in range(2, n):
        if n % i == 0:
            is_prime = False

    # 3) Print the result
    if is_prime:
        print("This is a prime number")
    else:
        print("This is not a prime number")
```

### Exercise 5

Code from lecture 1.2 and 2.1

```python
# Define the function
def find_min_max(numbers):
    min_value = None
    max_value = None

    for number in numbers:
        if min_value is None or number < min_value:
            min_value = number

        if max_value is None or number > max_value:
            max_value = number

    return min_value, max_value


# Pass one collection as a single argument
values = [4, 9, 1, 7, 3]

# Unpack the returned values
min_value, max_value = find_min_max(values)

# Print the results
print(f"Min value is: {min_value}")
print(f"Max value is: {max_value}")
```

```python
import random

# Choose a random number between 1 and 10
secret_number = random.randint(1, 10)

# Ask the user to guess the number
guess = int(input("Guess the number between 1 and 10: "))

# Keep asking until the guess is correct
while guess != secret_number:
    if guess > secret_number:
        print("Too high! Try again.")
    else:
        print("Too low! Try again.")

    guess = int(input("Guess the number between 1 and 10: "))

# Print success message
print("You guessed the right number.")
```

### Exercise 8

Code from lecture 1.2 and 2.1

```python
import math

def series(x, n):
    s = 0

    for i in range(1, n + 1):
        s = s + x * (5 ** i) ** 2 / (i + math.factorial(i + 1))

    return s

x = float(input("Enter x: "))
n = int(input("Enter N: "))

answer = series(x, n)
print(answer)
```

```python
import numpy as np
import matplotlib.pyplot as plt  # displaying images
import PIL  # Python Imaging Library - image editing
import urllib  # Library handling URLs (like requesting)
```

```python
image_url = "https://thumbs.dreamstime.com/b/ocean-sunset-landscape-bird-high-resolution-image-flying-towards-colorful-romantic-sky-246711534.jpg"
```

### Exercise 10

Code from lecture 2.2 and 3.1 

```python
import numpy as np
```

```python
# 1) Create a NumPy array
temperatures = np.array([22, 24, 19, 21, 23, 25, 26, 27, 24, 23,
                         20, 22, 21, 19, 18, 22, 23, 24, 26, 28,
                         30, 27, 25, 26, 24, 22, 21, 20, 19, 23])
```

```python
# 2) Find minimum, maximum, and average temperature
print(f"Minimum temperature is: {temperatures.min()}")
print(f"Maximum temperature is: {temperatures.max()}")
print(f"Average temperature is: {temperatures.mean()}")
```

```python
# 3) Convert Celsius to Fahrenheit
fahrenheit = temperatures * 9 / 5 + 32
print("Temperatures in Fahrenheit are:")
print(fahrenheit)
```

```python
# 4) Find all days where temperature was above 25°C
days_above_25 = np.where(temperatures > 25)[0] + 1
print("Days with temperature above 25°C are:")
print(days_above_25)
```

```python
# 5) Count how many days had temperature below 20°C
count_below_20 = np.sum(temperatures < 20)
print(f"Number of days below 20°C is: {count_below_20}")
```

```python
# 6) Sort temperatures in ascending order
sorted_temperatures = np.sort(temperatures)
print("Sorted temperatures are:")
print(sorted_temperatures)
```

```python
# 7) Reshape into 5x6 array
weeks = temperatures.reshape(5, 6)
print("5x6 reshaped array is:")
print(weeks)
```
