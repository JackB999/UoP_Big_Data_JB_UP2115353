### Exercise 1

Code from lecture 1.2 - 2.1

```python
# 1) Ask user to enter a number
n = int(input("Enter a number: "))

# 2) Check if the number is prime
if n <= 1:
    print("This is not a prime number")
else:
    is_prime = True

    for i in range(2, n):
        if n % i == 0:
            is_prime = False

    # 3) Print the result
    if is_prime:
        print("This is a prime number")
    else:
        print("This is not a prime number")
```

### Exercise 2

Code from lecture 2.1

```python
# Ask user how many Fibonacci numbers to display
n = int(input("How many Fibonacci numbers? "))

a = 0
b = 1

# Print Fibonacci sequence
for i in range(n):
    print(a, end=' ')
    next_num = a + b
    a = b
    b = next_num
```

### Exercise 3

Code from lecture 2.1 - Exercise 2

```python
n = 5

# top
for i in range(n, 0, -1):
    for j in range(i):
        print('*', end=' ')
    print()

# bottom
for i in range(2, n + 1):
    for j in range(i):
        print('*', end=' ')
    print()
```

```python
# Define the function
def reverse_text():
    text = input("Enter a string: ")
    
    reversed_text = ""
    
    for letter in text:
        reversed_text = letter + reversed_text
    
    print(reversed_text)

# Call the function
reverse_text()
```

### Exercise 5

Code from lecture 1.2 and 2.1

```python
# Define the function
def find_min_max(numbers):
    min_value = None
    max_value = None

    for number in numbers:
        if min_value is None or number < min_value:
            min_value = number

        if max_value is None or number > max_value:
            max_value = number

    return min_value, max_value


# Pass one collection as a single argument
values = [4, 9, 1, 7, 3]

# Unpack the returned values
min_value, max_value = find_min_max(values)

# Print the results
print(f"Min value is: {min_value}")
print(f"Max value is: {max_value}")
```

### Exercise 6

Code from lecture 1.2 and 2.1

```python
import random

# Choose a random number between 1 and 10
secret_number = random.randint(1, 10)

# Ask the user to guess the number
guess = int(input("Guess the number between 1 and 10: "))

# Keep asking until the guess is correct
while guess != secret_number:
    if guess > secret_number:
        print("Too high! Try again.")
    else:
        print("Too low! Try again.")

    guess = int(input("Guess the number between 1 and 10: "))

# Print success message
print("You guessed the right number.")
```
